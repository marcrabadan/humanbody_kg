### Importing libraries

```python
from neo4j import GraphDatabase, RoutingControl
```

#### Authentication

```python
URI = "neo4j://localhost:7687"
AUTH = ("neo4j", "Neo4j2420!")
```

#### Useful functions

```python
def create_node(driver, node_name, id, name, function):
    driver.execute_query(
        f"MERGE ({id}:{node_name}" + " { id: $id, name: $name, function: $function})",
        id=id, name=name, function=function, database_="neo4j",
    )
    
def create_relationships(driver, source_id, source_entity, rel_type, target_id, target_entity):
    driver.execute_query(
        f"MATCH (a:{source_entity} " + " { id: $source_id }), " + f"(b:{target_entity}" + " { id: $target_id })"
        f"MERGE (a)-[:{rel_type}]->(b)", source_id=source_id, target_id=target_id, database_="neo4j",
    )
```

#### Creating all nodes and relationships

```python
driver = GraphDatabase.driver(URI, auth=AUTH)
```

#### Creating all nodes

```python
for system in systems:
    create_node(driver, "System", system["id"], system["name"], system["function"])
    
for organ in organs:
    create_node(driver, "Organ", organ["id"], organ["name"], organ["function"])
    
for cell in cells:
    create_node(driver, "Cell", cell["id"], cell["name"], cell["function"])
    
for molecule in molecules:
    create_node(driver, "Molecule", molecule["id"], molecule["name"], molecule["function"])
    
for tissue in tissues:
    create_node(driver, "Tissue", tissue["id"], tissue["name"], tissue["function"])
    
for substance in substances:
    create_node(driver, "Substance", substance["id"], substance["name"], substance["function"])
    
for condition in conditions:
    create_node(driver, "Condition", condition["id"], condition["name"], condition["function"])
    
for disease in diseases:
    create_node(driver, "Disease", disease["id"], disease["name"], disease["function"])
    
for gene in genes:
    create_node(driver, "Gene", gene["id"], gene["name"], gene["function"])
    
for mutation in mutations:
    create_node(driver, "Mutation", mutation["id"], mutation["name"], mutation["function"])
    
for treatment in treatments:
    create_node(driver, "Treatment", treatment["id"], treatment["name"], treatment["function"])
    
for factor in factors:
    create_node(driver, "Factor", factor["id"], factor["name"], factor["function"])
    
for body_node in body_nodes:
    create_node(driver, "Body", body_node["id"], body_node["name"], body_node["function"])
    
    
```

#### Creating all relationships

```python
for relationship in relationships:
    create_relationships(driver, relationship["source_id"], relationship["source_entity"], relationship["rel_type"], relationship["target_id"], relationship["target_entity"])
```
